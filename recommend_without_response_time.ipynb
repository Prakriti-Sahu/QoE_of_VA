# **Importing Data and Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from warnings import filterwarnings
filterwarnings(action='ignore')
```

```python
cols=['ID', 'start_time', 'completion_time', 'email', 'name', 'name_2', 'age', 'gender',
          'occupation', 'education', 'skills', 'device', 'va', 'ease_1', 'response_time_1', 
          'ease_3', 'relevance_3', 'ease_2', 'response_time_2', 'relevance_4', 'response_time_4', 
          'relevance_5', 'response_time_5', 'clarity_7', 'relevance_6', 'response_time_6', 
          'relevance_7', 'response_time_7', 'clarity_8', 'performance', 'recommend', 'response_time_3']
```

```python
data = pd.read_csv('/content/English - English.csv', names = cols, index_col=False, header = None)
```

```python
data.head(5)
```

```python
data=data.iloc[1: , :]
data.head(5)
```

```python
data.columns
```

```python
x=data.drop(['ID','start_time', 'completion_time', 'email', 'name', 'name_2','performance', 'recommend', 
             'response_time_1', 'response_time_2', 'response_time_3', 'response_time_4', 'response_time_5',
             'response_time_6', 'response_time_7'],axis=1)
x.head(5)
```

```python
y1=data[['performance']]
y2=data[['recommend']]
print(y1.head(2))
print(y2.head(2))
```

# **Data Preprocessing**





```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
x['occupation']=le.fit_transform(x['occupation'])
x['gender']=le.fit_transform(x['gender'])
x['education']=le.fit_transform(x['education'])
x['device']=le.fit_transform(x['device'])
x['va']=le.fit_transform(x['va'])
```

```python
for i in x.columns:
  x[i]=x[i].astype(int)
```

```python
y1['performance'] = y1['performance'].astype(int)
y2['recommend'] = y2['recommend'].astype(int)
```

```python
y2.head()
```

```python
len(y2)
```

```python
y2['recommend'].value_counts()
```

```python
for i in range(1,len(y2)+1):
  if(y2["recommend"][i]>=7):
    y2["recommend"][i]=1
  else:
    y2["recommend"][i]=0

y2.head()
```

# **Training and Testing**

```python
from sklearn.model_selection import train_test_split
xtrain2, xtest2, ytrain2, ytest2 = train_test_split(x, y2, test_size=0.3,stratify=y2,random_state=42)
```

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import  accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression(random_state = 0)
model1.fit(xtrain2, ytrain2)
ypred_1 = model1.predict(xtest2)
y_train_pred=model1.predict(xtrain2)
print("Train Accuracy: ", accuracy_score(y_train_pred,ytrain2))
print("Test Accuracy: ", accuracy_score(ytest2,ypred_1))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier(n_neighbors=5)
model2.fit(xtrain2,ytrain2)
y_train_pred=model2.predict(xtrain2)
y_pred2 = model2.predict(xtest2)
print("Train Accuracy:",accuracy_score(y_train_pred,ytrain2))
print("Test Accuracy:",accuracy_score(ytest2,y_pred2))
```

```python
from sklearn.naive_bayes import GaussianNB
model3 = GaussianNB()
model3.fit(xtrain2,ytrain2)
y_train_pred=model3.predict(xtrain2)
y_pred2 = model3.predict(xtest2)
print("Train Accuracy:",accuracy_score(y_train_pred,ytrain2))
print("Test Accuracy:",accuracy_score(ytest2,y_pred2))
```

```python
from sklearn.tree import DecisionTreeClassifier
model4 = DecisionTreeClassifier(criterion='entropy',random_state=7)
model4.fit(xtrain2,ytrain2)
y_train_pred=model4.predict(xtrain2)
y_pred2 = model4.predict(xtest2)
print("Train Accuracy:",accuracy_score(y_train_pred,ytrain2))
print("Test Accuracy:",accuracy_score(ytest2,y_pred2))
```

```python
from sklearn.svm import SVC
model5 = SVC()
model5.fit(xtrain2, ytrain2)
ypred_1=model5.predict(xtest2)
y_train_pred=model5.predict(xtrain2)
print("Train Accuracy: ", accuracy_score(y_train_pred,ytrain2))
print("Test Accuracy: ", accuracy_score(ytest2,ypred_1))
```

```python
from sklearn.ensemble import RandomForestClassifier
model6= RandomForestClassifier(n_estimators = 100) 
model6.fit(xtrain2, ytrain2)
ypred_1=model6.predict(xtest2)
y_train_pred=model6.predict(xtrain2)
print("Train Accuracy: ", accuracy_score(y_train_pred,ytrain2))
print("Test Accuracy: ", accuracy_score(ytest2,ypred_1))
```
